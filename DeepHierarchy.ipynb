```python
import cv2
import os
from tqdm import tqdm
Data=list()
def create_data(directory_name,lab):
    for filename in tqdm(os.listdir(directory_name)):
        img=cv2.imread(os.path.join(directory_name,filename))
        try:
            img=cv2.resize(img,(150,150))
            Data.append([img,lab])
        except:
            pass
create_data("Covid_Data_GradientCrescent\\all\\train\\covid",0)
create_data("Covid_Data_GradientCrescent\\all\\train\\pneumonia_bac",1)
create_data("Covid_Data_GradientCrescent\\all\\train\\pneumonia_vir",2)
print(len(Data))
```

```python
Val=list()
def create_data_test(directory_name,lab):
    for filename in tqdm(os.listdir(directory_name)):
        img=cv2.imread(os.path.join(directory_name,filename))
        try:
            img=cv2.resize(img,(150,150))
            Val.append([img,lab])
        except:
            pass
create_data_test("Covid_Data_GradientCrescent\\all\\test\\covid",0)
create_data_test("Covid_Data_GradientCrescent\\all\\test\\pneumonia_bac",1)
create_data_test("Covid_Data_GradientCrescent\\all\\test\\pneumonia_vir",2)
print(len(Val))
```

```python
import random
random.shuffle(Data)
random.shuffle(Val)
X_train=[]
Y_train=[]

X_test=[]
Y_test=[]


for img,lab in Data:
    X_train.append(img)
    Y_train.append(lab)
    
for img,lab in Val:
    X_test.append(img)
    Y_test.append(lab)
    
```

```python
import numpy as np
X_train=np.array(X_train)
X_train=X_train.astype('float32')/255
X_test=np.array(X_test)
X_test=X_test.astype('float32')/255
```

```python
import keras
from keras import models
from keras import layers
from keras.applications import VGG19
from keras import optimizers
from keras.layers.core import Flatten, Dense, Dropout, Lambda
conv_base = VGG19(weights='imagenet',include_top=False,input_shape=(150, 150, 3))

conv_base.trainable = False
model = models.Sequential()
model.add(conv_base)
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(3, activation='softmax'))
model.compile(loss='sparse_categorical_crossentropy',optimizer=optimizers.Adam(lr=0.0005),metrics=['acc'])
print(model.summary)
```

```python
from keras.callbacks import ModelCheckpoint, EarlyStopping
checkpoint = ModelCheckpoint("second_level_hierarchy.h5", monitor='val_acc', verbose=1, save_best_only=True, save_weights_only=False, mode='auto', period=1)
early = EarlyStopping(monitor='val_accuracy', min_delta=0, patience=10, verbose=1, mode='auto')
model.fit(X_train,Y_train,epochs=10,validation_data=(X_test,Y_test),batch_size=5,callbacks=[checkpoint,early])
model.save("second_level_hierarchy.h5")
```
